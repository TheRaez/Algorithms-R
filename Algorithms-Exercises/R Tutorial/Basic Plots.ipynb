# 5.1 Strip Charts

```r
help(stripchart)
```

```r
stripchart(w1)
stripchart(w1, method='stack')
stripchart(w1, method='jitter')
```

```r
stripchart(w1, vertical=TRUE)
stripchart(w1, vertical=TRUE, method="jitter")
```

```r
stripchart(w1, method='stack',
           main='Leaf BioMass in High CO2 Environment',
          xlab="BioMass of Leaves")
```

```r
stripchart(w1, method='jitter', vertical=TRUE)
title('LeafBioMass in High CO2 Environment', ylab='BioMass of Leaves')
```

# 5.2 Histograms

```r
help(hist)
```

```r
hist(w1$vals)
hist(w1$vals, main='Distribution of w1',xlab="w1")
```

```r
hist(w1$vals, breaks=2)
hist(w1$vals, breaks=6)
hist(w1$vals, breaks=12)
```

```r
hist(w1$vals, breaks=12, xlim=c(0,10))
hist(w1$vals, breaks=12, xlim=c(-1,2))
hist(w1$vals, breaks=12, xlim=c(0,2))
hist(w1$vals, breaks=12, xlim=c(1,1.3))
hist(w1$vals, breaks=12, xlim=c(0.9,1.3))
```

```r
hist(w1$vals,
     main='Leaf BioMass in High CO2 Environment',
     xlab='BioMass of Leaves')
```

```r
hist(w1$vals)
title('Leaf BioMass in High CO2 Environment',xlab='BioMass of Leaves')
```

```r
hist(w1$vals,main='Leaf BioMass in High CO2 Environment',xlab='BioMass of Leaves',ylim=c(0,16))
stripchart(w1$vals,method='stack',add=TRUE,at=15.5)
```

# 5.3 Boxplots

```r
help(boxplot)
```

```r
boxplot(w1,
        main='Leaf BioMass in High CO2 Environment',
        ylab='BioMass of Leaves')
```

```r
boxplot(w1$vals,
        main='Leaf BioMass in High CO2 Environment',
        xlab='BioMass of Leaves',
        horizontal=TRUE)
```

```r
hist(w1$vals,
     main='Leaf BioMass in High CO2 Environment',
     xlab='BioMass of Leaves',
     ylim=c(0,16))
boxplot(w1$vals,
        horizontal=TRUE,
        at=15.5,
        add=TRUE,
        axes=FALSE)
```

```r
hist(w1$vals,main='Leaf BioMass in High CO2 Environment',xlab='BioMass of Leaves',ylim=c(0,16))
boxplot(w1,horizontal=TRUE,at=16,add=TRUE,axes=FALSE)
stripchart(w1,add=TRUE,method='stack',at=15)
```
