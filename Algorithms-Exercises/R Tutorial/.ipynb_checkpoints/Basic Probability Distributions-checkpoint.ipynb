4.1 The Normal Distribution

```r
help(Normal)
```

```r
dnorm(0)
dnorm(0)*sqrt(2*pi)
dnorm(0, mean=4)
dnorm(0, mean=4, sd=10)
```

```r
x <- seq(-20, 20, by=.01)
y <- dnorm(x)
plot(x,y)
y <- dnorm(x, mean=2,5, sd=0.01)
plot(x,y)
```

```r
pnorm(0, lower.tail=FALSE)
pnorm(1, lower.tail=FALSE)
pnorm(0, mean=2, lower.tail=FALSE)
x <- seq(-20, 20, by=.01)
y <- pnorm(x)
plot(x, y)
```

```r
qnorm(0.5)
qnorm(0.5, mean=1)
qnorm(0.5, sd=10)
```

```r
v <- c(0.1, 0.3, 0.75)
qnorm(v)
```

```r
x <- seq(0, 1, by=0.01)
y <- qnorm(x)
plot(x, y)
y <- qnorm(x, mean=4)
plot(x, y)
y <- qnorm(x, sd=10)
plot(x, y)
```

```r
rnorm(4)
rnorm(4, mean=3)
```

```r
y <- rnorm(300, mean=3)
hist(y)
qqnorm(y)
qqline(y)
```

4.2 The t Distribution

```r
help(TDist)
```

```r
x <- seq (-20, 20, by=.05)
y <- dt(x, df=10)
plot(x, y)
y <- dt(x, df=50)
plot(x, y)
```

```r
x <- seq(-20, 20, by=.05)
y <- pt(x, df=10)
plot(x, y)
```

```r
x <- seq(0, 1, by=0.01)
y <- qt(x, df=10)
plot(x, y)
```

```r
y = rt(300, df=10)
hist(y)
```

4.3 The Binomial Distribution

```r
help(Binomial)
```

```r
x <- seq(0, 500, by=1)
y <- dbinom(x, 500, .2)
plot(x, y)
y <- dbinom(x, 500, .6)
plot(x, y)
```

```r
y <- pbinom(x, 500, 0.6)
plot(x, y)
```

```r
qbinom(0.5, 500, 0.6)
```

```r
y <- rbinom(500, 1000, .2)
hist(y)
```

4.4 The Chi-Squared Distribution

```r
help(Chisquare)
```

```r
x <- seq(-20, 20, by=.05)
y <- dchisq(x, df=8)
plot(x, y)
y <- dchisq(x, df=16)
plot(x, y)
```

```r
x <- c(2,4,5,6)
pchisq(x, df=8)
```

```r
qchisq(0.05, df=12)
```

```r
rchisq(10, df=10)
```
